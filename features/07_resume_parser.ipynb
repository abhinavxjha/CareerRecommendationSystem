```python
import pdfplumber as pf
import json
```

```python
def text_from_pdf(filename):    
    with pf.open(filename) as pdf:
        text=""
        for i in pdf.pages:
            page_text=i.extract_text()
            if page_text:
                text += page_text + "\n"
    return text
```

```python
d=text_from_pdf("../data/AbhinavJhaResume.pdf")
skills_data=load_json("../data/skills.json")
```

```python
all_skills=[]
for category in skills_data:
    all_skills.extend(skills_data[category])
```

```python
def extract_skills(d, all_skills):
    common_skills=[]
    for skill in all_skills:
        if skill.lower() in d.lower():
            common_skills.append(skill)
    return common_skills
```

```python
user_skills=extract_skills(d, all_skills)
```
