```python
import sys
sys.path.append("../modules")
from helpers import (
    load_json,
    get_required_skills,
    career_selection,
    values,
    course_recommendation,
    load_courses
)
```

```python
required_skills_data=load_json("../data/careers.json")
courses_data=load_courses('../data/courses.json')
selected_career=career_selection()
skills=input("Enter Your Skills :: ")
user_skills=skills.split()
```

```python
mskills=values(required_skills_data, selected_career, user_skills)
courses=course_recommendation(mskills, courses_data)
```

```python
def roadmap_generator(mskills, courses):
    roadmap={}
    for month, skill in enumerate(mskills, start=1):
        roadmap[f"Month {month}"]={"skill":skill}
    return roadmap
```

```python
final_roadmap=roadmap_generator(mskills, courses)
for i,j in final_roadmap.items():
    print(i ,"-", j)
    
    print("-" * 30)
```
