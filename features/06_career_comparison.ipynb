```python
import pandas as pd
import json
```

```python
import sys
sys.path.append("../modules")

from helpers import (
    load_json,
    matching_score,
    get_career_info,
    load_career,
    career_selection,
    skills_input
)
```

```python
career_file=load_career("../data/career_info.json")
required_skills_data=load_json("../data/careers.json")
user_skills=skills_input()
results=matching_score(required_skills_data, user_skills)
```

```python
comparison_rows=[]
for career,score in results:
    info=get_career_info(career_file, career)
    if score > 0:
        if info is None:
            continue
        row = {"Career": career, "Match Score": score, "Difficulty": info["difficulty"], "Learning Time": info["learning_time"]}
        comparison_rows.append(row)
```

```python
df=pd.DataFrame(comparison_rows)
df["Match Score"] = df["Match Score"].round(2)
df=df.sort_values(by="Match Score", ascending=False)
```
